```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans,AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
df = pd.read_csv('Mall_Customers.csv')
df
```

```python
df.isnull().sum()
```

```python
df.drop('CustomerID',axis=1,inplace=True)
```

```python
X = df.iloc[:,2:]
X
```

## Detecting outliers

```python
X.boxplot()
```

## Elbow method

```python
wcss = []           
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(X)
    print(f'Cost_Function={kmeans.inertia_} with {i} Clusters')
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

5 clusters are the best no. of clusters that we will use

## KMean

```python
model = KMeans(n_clusters=5,n_init=50, random_state=0)
y_kmeans= model.fit_predict(X)
```

```python
def plotting(y):
    for i in range(5):
        plt.scatter(np.array(X)[y == i,0], np.array(X)[y == i,1], label=f'Cluster {i + 1}')
    plt.title('Clusters of customers')
    plt.xlabel('Annual Income')
    plt.ylabel('Spending Score')
    plt.legend(loc = 'best')
    plt.show()
```

```python
plotting(y_kmeans)
```

## Hierachical clustering

```python
def dendogramplot(method):
    dendrogram = sch.dendrogram(sch.linkage(X, method  = method))
    plt.title('Dendrogram')
    plt.xlabel('Customers')
    plt.ylabel('Euclidean distances')

    plt.show()
```

```python
dendogramplot('ward')
```

```python
hc = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc = hc.fit_predict(X)
plotting(y_hc)
```

```python
dendogramplot('single')
```

```python
hc = AgglomerativeClustering(n_clusters=5,affinity='cosine',linkage='single')
y_hc = hc.fit_predict(X)
plotting(y_hc)
```

```python
dendogramplot('average')
```

```python
hc = AgglomerativeClustering(n_clusters=5,affinity='manhattan',linkage='average')
y_hc = hc.fit_predict(X)
plotting(y_hc)
```

```python
dendogramplot('complete')
```

```python
hc = AgglomerativeClustering(n_clusters=5,affinity='l1',linkage='complete')
y_hc = hc.fit_predict(X)
plotting(y_hc)
```

```python
db = DBSCAN(eps=0.4, min_samples=5).fit(X)
labels = db.labels_
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)
print(f'no. of clusters {n_clusters}')
print(f'noise is {n_noise}')
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
```

```python
unique_labels = set(labels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=14)

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters)
plt.show()
```
